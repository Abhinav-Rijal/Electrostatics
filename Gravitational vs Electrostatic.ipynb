```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class Coulomb():
    def __init__(self,k,q1,q2,r):
        self.k=k
        self.q1=q1
        self.q2=q2
        self.r=r
    def Force(self):
        return self.k*self.q1*self.q2/self.r**2
    def Intensity(self):
        return self.k*self.q1/self.r**2
    def Potential(self):
        return self.k*self.q1/self.r
    def update_r(self,r):
        self.r=r
    def update_charge(self,q1,q2):
        self.q1=q1
        self.q2=q2
```

```python
C=Coulomb(9*10e9,0,0,0)
C.update_charge(3.2e-6,3.2e-10)
#print(C.test_q1())
R=[] #empty list to collect distances
F=[] #empty list to collect forces
E=[] #empty list to collect intensities
V=[] #empty list to collect potentials
for r in range(1,1000):
    C.update_r(r)
    f=C.Force()
    e=C.Intensity()
    v=C.Potential()
    R.append(r)
    F.append(f)
    E.append(e)
    V.append(v)
    
plt.figure(figsize=[9,9])

plt.subplot(2,2,1)
plt.plot(R,F,label='force')
plt.xlabel('r')
plt.ylabel('F,E,V')
plt.plot(R,E,label='intensity')
plt.plot(R,V,label='potential')
plt.title('Normal scale')
plt.legend()

plt.subplot(2,2,2)
plt.semilogy(R,F,label='force')
plt.xlabel('r')
plt.ylabel('F,E,V')
plt.semilogy(R,E,label='intensity')
plt.semilogy(R,V,label='potential')
plt.title('Semilog scale')
plt.legend()

plt.subplot(2,2,3)
plt.loglog(R,F,label='force')
plt.xlabel('r')
plt.ylabel('F,E,V')
plt.loglog(R,E,label='intensity')
plt.loglog(R,V,label='potential')
plt.title('loglog scale')
plt.legend()
```

```python
class Gravitation():
    def __init__(self,G,m1,m2,r):
        self.G=G
        self.m1=m1
        self.m2=m2
        self.r=r
    def Force(self):
        return self.G*self.m1*self.m2/self.r**2
    def Intensity(self):
        return self.G*self.m1/self.r**2
    def Potential(self):
        return self.G*self.m1/self.r
    def update_r(self,r):
        self.r=r
    def update_mass(self,m1,m2):
        self.m1=m1
        self.m2=m2
```

```python
G=Gravitation(6.67*10e-11,0,0,0)
G.update_mass(5,10)
#print(C.test_q1())
R=[] #empty list to collect distances
F=[] #empty list to collect forces
E=[] #empty list to collect intensities
V=[] #empty list to collect potentials
for r in range(1,1000):
    G.update_r(r)
    f=G.Force()
    e=G.Intensity()
    v=G.Potential()
    R.append(r)
    F.append(f)
    E.append(e)
    V.append(v)
    
plt.figure(figsize=[9,9])

plt.subplot(2,2,1)
plt.plot(R,F,label='force')
plt.xlabel('r')
plt.ylabel('F,E,V')
plt.plot(R,E,label='intensity')
plt.plot(R,V,label='potential')
plt.title('Normal scale')
plt.legend()

plt.subplot(2,2,2)
plt.semilogy(R,F,label='force')
plt.xlabel('r')
plt.ylabel('F,E,V')
plt.semilogy(R,E,label='intensity')
plt.semilogy(R,V,label='potential')
plt.title('Semilog scale')
plt.legend()

plt.subplot(2,2,3)
plt.loglog(R,F,label='force')
plt.xlabel('r')
plt.ylabel('F,E,V')
plt.loglog(R,E,label='intensity')
plt.loglog(R,V,label='potential')
plt.title('loglog scale')
plt.legend()
```

```python
#for H-atom
C=Coulomb(9*10e9,0,0,5.29*10e-11)
C.update_charge(1.6*10e-16,1.6e10-11)
G=Gravitation(6.67*10e-11,0,0,5.29*10e-11)
G.update_mass(9.1*10e-31,1.67e10-27)

fe=C.Force()
fg=G.Force()
print(fe),print(fg)
if fe>fg:
    print('Electrostatic force is greater than Gravitational force between an electron and proton in H-atom by', fe/fg, 'times.')
elif fg>fe:
    print('Gravitational force is greater than Electrostatic force between an electron and proton in H-atom by', fg/fe, 'times.')
else:
    print('Gravitational force is equal to Electrostatic force between an electron and proton in H-atom')
```

```python
#for a range of distances
R=[] #empty list to collect distances
Fe=[] #empty list to collect electrostatic force
Fg=[] #empty list to collect gravitational force

for r in range(1,1000):
    G.update_r(r)
    C.update_r(r)
    fg=G.Force()
    fe=C.Force()
    
    R.append(r)
    Fe.append(fe)
    Fg.append(fg)
    
    
plt.figure(figsize=[9,9])

plt.subplot(2,2,1)
plt.plot(R,Fe,label='Electrostatic force')
plt.xlabel('r')
plt.ylabel('G vs E force')
plt.plot(R,Fg,label='Gravitational force')
plt.title('Normal scale')
plt.legend()

plt.subplot(2,2,2)
plt.semilogy(R,Fe,label='Electrostatic force')
plt.xlabel('r')
plt.ylabel('G vs E force')
plt.semilogy(R,Fg,label='Gravitational force')
plt.title('Semilog scale')
plt.legend()

plt.subplot(2,2,3)
plt.loglog(R,Fe,label='Electrostatic force')
plt.xlabel('r')
plt.ylabel('G vs E force')
plt.loglog(R,Fg,label='Gravitational force')
plt.title('Loglog scale')
plt.legend()
```
