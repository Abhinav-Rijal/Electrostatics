Coulomb's force between two point charged particles is given by:
$$ \vec{F}=k\frac{q_1q_2}{r^2}\hat{r} $$ 

where;

$q_1$= point charge taken at origin '$O$'

$q_2$= point charge(test charge) taken at a distance '$r$' from $q_1$

$\vec{r}$= position vector of $q_2$

$\hat{r}$= $\frac{\vec{r}}{r}$ = unit vector(outward radial) along the position vector of $q_2$

and the magnitude of the force is;
$$ F= k\frac{q_1q_2}{r^2} $$

The electric field intensity of pt. charge '$q_1$' at a point at distance '$r$' from it is defined as;

$$ \vec{E}=\frac{\vec{F}}{q_2}=k\frac{q_1}{r^2}\hat{r} $$ 

Again, the magnitude of electric field intensity is;
$$ E=k\frac{q_1}{r^2} $$

And the electric potential of pt. charge '$q_1$' at a point at distance '$r$' from it is;
$$ V=k\frac{q_1}{r} $$

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class Coulomb():
    def __init__(self,k,q1,q2):
        self.k=k
        self.q1=q1
        self.q2=q2
    def Force(self,r):
        return self.k*self.q1*self.q2/r**2
    def Intensity(self,r):
        return self.k*self.q1/r**2
    def Potential(self,r):
        return self.k*self.q1/r
    def test_q1(self):
        return self.q1
    def update_charge(self,q1,q2):
        self.q1=q1
        self.q2=q2
```

```python
C1=Coulomb(k=9*10e9, q1=10e-6, q2=10e-4)
f=C1.Force(10)
e=C1.Intensity(10)
v=C1.Potential(10)
print('The force, intensity and potential are:')
print(f),print(e),print(v)
```

```python
C=Coulomb(9*10e9,0,0) #initialize object C of class Coulomb with arguments 0,0,0
print(C.test_q1())
q1=9 #doesn't take this value
print(C.test_q1())
```

```python
#to update data of class(q1,q2), we have defined update function in Coulomb
C.update_charge(9,9)
print(C.test_q1())
```

```python
C.update_charge(3.2e-6,3.2e-10)
#print(C.test_q1())
R=[] #empty list to collect distances
F=[] #empty list to collect forces
E=[] #empty list to collect intensities
V=[] #empty list to collect potentials
for r in range(1,1000):
    f=C.Force(r)
    e=C.Intensity(r)
    v=C.Potential(r)
    R.append(r)
    F.append(f)
    E.append(e)
    V.append(v)
    
plt.figure(figsize=[9,9])

plt.subplot(2,2,1)
plt.plot(R,F,label='force')
plt.xlabel('r')
plt.ylabel('F,E,V')
plt.plot(R,E,label='intensity')
plt.plot(R,V,label='potential')
plt.title('Normal scale')
plt.legend()

plt.subplot(2,2,2)
plt.semilogy(R,F,label='force')
plt.xlabel('r')
plt.ylabel('F,E,V')
plt.semilogy(R,E,label='intensity')
plt.semilogy(R,V,label='potential')
plt.title('Semilog scale')
plt.legend()

plt.subplot(2,2,3)
plt.loglog(R,F,label='force')
plt.xlabel('r')
plt.ylabel('F,E,V')
plt.loglog(R,E,label='intensity')
plt.loglog(R,V,label='potential')
plt.title('loglog scale')
plt.legend()
```

```python
#making r as class variable
class Coulomb():
    def __init__(self,k,q1,q2,r):
        self.k=k
        self.q1=q1
        self.q2=q2
        self.r=r
    def Force(self):
        return self.k*self.q1*self.q2/self.r**2
    def Intensity(self):
        return self.k*self.q1/self.r**2
    def Potential(self):
        return self.k*self.q1/self.r
    def update_r(self,r):
        self.r=r
    def update_charge(self,q1,q2):
        self.q1=q1
        self.q2=q2
```

```python
C=Coulomb(9*10e9,0,0,0)
C.update_charge(3.2e-6,3.2e-10)
#print(C.test_q1())
R=[] #empty list to collect distances
F=[] #empty list to collect forces
E=[] #empty list to collect intensities
V=[] #empty list to collect potentials
for r in range(1,1000):
    C.update_r(r)
    f=C.Force()
    e=C.Intensity()
    v=C.Potential()
    R.append(r)
    F.append(f)
    E.append(e)
    V.append(v)
    
plt.figure(figsize=[9,9])

plt.subplot(2,2,1)
plt.plot(R,F,label='force')
plt.xlabel('r')
plt.ylabel('F,E,V')
plt.plot(R,E,label='intensity')
plt.plot(R,V,label='potential')
plt.title('Normal scale')
plt.legend()

plt.subplot(2,2,2)
plt.semilogy(R,F,label='force')
plt.xlabel('r')
plt.ylabel('F,E,V')
plt.semilogy(R,E,label='intensity')
plt.semilogy(R,V,label='potential')
plt.title('Semilog scale')
plt.legend()

plt.subplot(2,2,3)
plt.loglog(R,F,label='force')
plt.xlabel('r')
plt.ylabel('F,E,V')
plt.loglog(R,E,label='intensity')
plt.loglog(R,V,label='potential')
plt.title('loglog scale')
plt.legend()
```
